```python
import pandas as p
import re #text manipulation
from nltk.corpus import stopwords #stopwords.words('english')
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Master Functions
def clean_corpus_series(corpus_series, remove_stop_words = True, custom_stop_words = []):
    for i in range(0, len(corpus_series)):
        corpus_series[i] = clean_string_many_words(corpus_series[i], remove_stop_words, custom_stop_words)
    return corpus_series
def document_term_matrix(corpus_series):
    countvectorizer = CountVectorizer()
    data_countvectorizer = countvectorizer.fit_transform(corpus_series)
    document_term_matrix1 = p.DataFrame(
        data_countvectorizer.toarray(),
        columns = countvectorizer.get_feature_names()
    )
    return(document_term_matrix1)
```

```python
def clean_string_many_words(string_many_words, remove_stop_words = True, custom_stop_words = []):
    string_many_words = string_many_words.replace('\n', " ")
    string_series = p.Series(
        re.split('[ -]', string_many_words)
    )
    #string_many_words.split(sep = " ")
    clean_function = lambda x: clean_word(x)
    string_series = string_series.apply(clean_function)
    if remove_stop_words == True:
        string_series = [word for word in string_series\
                         if word not in stopwords.words('english')\
                         and word not in custom_stop_words]
    return ' '.join(string_series)
def clean_word(word):
    if type(word) != str:
        word = ""
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    word = word.lower()
    word = re.sub('\[.*?\]', '', word)
    word = re.sub('[!"\\#\\$%\\&\'\\(\\)\\*\\+,\\./:;<=>\\?@\\[\\\\\\]\\^_`\\{\\|\\}\\~]', '', word)
    word = re.sub('\w*\d\w*', '', word)
    word = word.strip()
    return word
```

```python
# import data
data = p.read_pickle('data.pkl')
```

```python
# derive corpus series from data
corpus_series = data.text[0:100]
```

```python
# declare custom stop words
custom_stop_words = ['introduction', '']
```

```python
# clean corpus series
clean_corpus_series1 = clean_corpus_series(corpus_series, custom_stop_words = custom_stop_words)
clean_corpus_series1
```

```python
# create document term matrix
document_term_matrix1 = document_term_matrix(clean_corpus_series1)
document_term_matrix1
```

```python
# save document term matrix
document_term_matrix1.to_pickle('document_term_matrix.pkl')
```
