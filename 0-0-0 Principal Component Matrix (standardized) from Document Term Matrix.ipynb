```python
import pandas as p
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
document_term_matrix = p.read_pickle('document_term_matrix.pkl')
document_term_matrix
```

```python
document_term_matrix_standardized = p.DataFrame(
    scale(document_term_matrix),
    columns = document_term_matrix.columns
)
document_term_matrix_standardized
```

```python
# You may adjust the number of components in PCA() as n_components.
# Each row is a corpus.
# Each column is a principal component.
principal_component_matrix = p.DataFrame(
    PCA().fit_transform(
        document_term_matrix_standardized
    )
)
principal_component_matrix
```

```python
principal_component_matrix.to_pickle('principal_component_matrix.pkl')
# p.read_pickle('principal_component_matrix.pkl')
```
